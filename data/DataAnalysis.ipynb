```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

plt.style.use("relatorio")
```

### Overlap

```python
df_s  = pd.read_csv("DOPO overlap symmetric.csv")
df_as = pd.read_csv("DOPO overlap asymmetric.csv")
```

```python
f, ax = plt.subplots(figsize=(6, 6))

ax.plot(df_s.mu, df_s.f_DOPO)
ax.plot(df_as.mu, df_as.f_DOPO)

ax.set_ylim(0, 1)

ax.set_xlabel("$\\mu$")
ax.set_ylabel("$f_0^{-1, 0, 1}$")

f.savefig("DOPO overlap.svg")
```

### gain

```python
dfg_s  = pd.read_csv("gain.csv")
dfg_s["g"] = dfg_s["g"].str.replace(" + 0.i", "").astype("float")

dfg_as = pd.read_csv("gain_as.csv")
dfg_as["g"] = dfg_as["g"].str.replace(" + 0.i", "").astype("float")
```

```python
f, ax = plt.subplots(figsize=(6, 6))

ax.plot(dfg_s.mu, dfg_s.g)
ax.plot(dfg_as.mu, dfg_as.g)

# ax.set_ylim(0, 1)

ax.set_xlabel("$\\mu$")
ax.set_ylabel("gain")

# f.savefig("DOPO overlap.svg")
```

```python
f, ax = plt.subplots(1, 2, sharex=True, figsize=(9,4), constrained_layout=True)

# c_sym  = "#60b1adff"
# c_asym = "#1a5fb4ff"

c_sym  = "#5DC1B9"   # teal mais claro (mais brilho)
c_asym = "#0B3C8C"   # azul mais profundo (mais escuro)

c_sym  = "#6EC5B8"   # teal suave
c_asym = "#1F4E79"   # azul dessaturado escuro

ax[0].plot(df_s.mu, df_s.f_DOPO, color=c_sym, lw=2)
ax[0].plot(df_as.mu, df_as.f_DOPO, color=c_asym, lw=2)

ax[0].set_ylim(0, .65)

ax[0].set_xlabel("$\\mu$")
ax[0].set_ylabel("$f_0^{-1, 0, 1}$")



ax[1].plot(dfg_s.mu, dfg_s.g,  color=c_sym, lw=2)
ax[1].plot(dfg_as.mu, dfg_as.g,  color=c_asym, lw=2)

ax[1].set_ylim(ymin=-0.2)
lims = ax[1].get_xlim()

ax[1].hlines(0, *lims, "k", ls="--", alpha=0.7)
ax[1].text(18, 0.05, "Threshold", ha="right", va="center", fontsize=11)

ax[1].set_xlim(*lims)

ax[1].set_xlabel("$\\mu$")
ax[1].set_ylabel("gain")

ax[1].fill_between(dfg_s.mu, dfg_s.g, 0, where=(dfg_s.g > 0),
                   color=c_sym, alpha=0.15)

ax[1].fill_between(dfg_as.mu, dfg_as.g, 0, where=(dfg_as.g > 0),
                   color=c_asym, alpha=0.25)


ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()

# --- estilo ---
for a in ax:
    a.grid(True, alpha=0.2)

# f.tight_layout()
f.savefig("DOPO overlap and gain.svg")
```
